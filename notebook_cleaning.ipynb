## 3- Nettoyer les données
- Sélectionner les colonnes que je souhaite garder
- Valeurs aberrantes => nan
- Valeurs manquantes lambda => imputation rapide
- Supprimer les lignes trop peu remplies 
- Valeurs manquantes dans colonnes importantes => imputation knn

### Sélection colonnes

```python
# sélectionner les colonnes d'intérets en lien avec mon idée d'application
clean_cols = data.columns

for col in clean_cols:
    print('Taux de completion de la colonne ' + col)
    print('-'*5)
    taux = check_column_completion(data, col)
    print(taux)
    print('-'*40)
```

```python
valid_cols = [
    'product_name', 'generic_name', 'code', 'packaging', 'brands', 'origins_tags', 'nutrition_grade_fr',
    'manufacturing_places_tags', 'energy_100g', 'saturated-fat_100g', 'sugars_100g', 'proteins_100g', 'fiber_100g',
    'labels_fr', 'emb_codes_tags', 'pnns_groups_2', 'pnns_groups_1', 'image_url',
    'additives_n', 'ingredients_from_palm_oil_n', 'carbohydrates_100g','salt_100g', 'fat_100g'
]
data.drop(columns=data.columns.difference(valid_cols), inplace=True)
```

```python
# nombre de colonnes réduit à 23
data.shape
```

```python
for col in valid_cols:
    print('Valeurs unique de ' + col + ' :')
    print(data[col].unique())
    print('-'*20)
```

```python
# drop lignes dupliquées
data[data.duplicated(keep=False)].sort_values('code')
data.drop_duplicates(inplace=True)
```

```python
# détection de valeurs abérrantes
data.loc[data['energy_100g'] >= 3766, 'energy_100g'] = np.nan

col_nutri_100g_max = ['saturated-fat_100g', 'sugars_100g', 'salt_100g',
                      'proteins_100g', 'fiber_100g', 'carbohydrates_100g', 'fat_100g']
for col in col_nutri_100g_max:
    data.loc[data[col] > 100, col] = np.nan
    data.loc[data[col] < 0, col] = np.nan
```

```python
# valeurs manquantes - imputation

# supprimer les lignes où product et generic name sont nan
data = data[~(data['product_name'].isna()) & (data['generic_name'].isna())]

# image
data['image_url'].fillna(
    'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTyyh0dfqW6aN_gu70Kd2CCisMifNIRJjAG0g&usqp=CAU',
    inplace=True
)

# unknown pnns group
data['pnns_groups_1'].replace({"unknown": np.nan}, inplace=True)
data['pnns_groups_2'].replace({"unknown": np.nan}, inplace=True)

# emb_code et manufacturing place
data.loc[
     (data['emb_codes_tags'].notna()) & 
     (data['manufacturing_places_tags'].isna()) &
     (data['emb_codes_tags'].str[:3] == 'fr-') |
     (data['emb_codes_tags'].str.contains('emb-')),
     'manufacturing_places_tags'
    ] = 'France'

# variables non imputables
nan_to_unknown_cols = ['brands', 'labels_fr', 'origins_tags']
for nan_col in nan_to_unknown_cols:
    data.loc[data[nan_col].isna(), nan_col] = 'Inconnu'

# supprimer emb code
data.drop(columns='emb_codes_tags', inplace=True)
```

```python
for group in data['pnns_groups_1'].unique()[2:]:
    print(group)
    print(data[data['pnns_groups_1'] == group].describe(include='all').loc['count':'freq', 'packaging'])
    print('part du résultat', round(data[data['pnns_groups_1'] == group].describe(include='all').loc['freq', 'packaging'] * 100 /
         data[data['pnns_groups_1'] == group].describe(include='all').loc['count', 'packaging'], 2) )
    print('-'*20)
```

```python
# checker le taux de completion des lignes, après la mise à nan
nb_of_cols = len(data.columns)
data["row_completion"] = (data.notna().sum(1) * 100) // nb_of_cols
```

```python
# visuel row completion
plt.hist(data["row_completion"], bins=20, color='red', alpha=0.6)
```

```python
# garder les individus avec un taux de complétion de ligne supérieur ou égal à 55%
# 320k lignes => 215k lignes 

data = data[data['row_completion'] >= 55]
data.drop(columns=['row_completion'], inplace=True)
data.reset_index(inplace=True, drop=True)

# reindex image url to get cat variables / quantitative variables
image_col = data['image_url'].values
data.drop(columns=['image_url'], inplace=True)
image_col
data.insert(loc=6, column='image_url', value=image_col)
data.shape
```

```python
# parmi les variables qualitatives, trouver celles dont le nombre de valeurs uniques est faible
# et les  transformer avec one hot encoding ou dummy
# => pnns_groups_1', 'pnns_groups_2' 'nutrition_grade_fr'
qual_cols = ['pnns_groups_1', 'pnns_groups_2','manufacturing_places_tags',
             'packaging', 'origins_tags', 'nutrition_grade_fr']

for col in qual_cols:
    print(col)
    print(data[col].unique())
    print(len(data[col].unique()))
    print('_'*20)
```

```python
# imputation knn avec dummy variables
cat_dummies = pd.get_dummies(data, columns=[
    'pnns_groups_1', 'pnns_groups_2', 'nutrition_grade_fr'
])
```

```python
# encoder 
X = cat_dummies.loc[:,"additives_n":]
data_not_imputed = data.loc[:, :"labels_fr"]

# checker les nan pour group 1 et 2, nutrition grade
X.loc[data.pnns_groups_1.isnull(), X.columns.str.startswith("pnns_groups_1_")] = np.nan
X.loc[data.pnns_groups_2.isnull(), X.columns.str.startswith("pnns_groups_2_")] = np.nan
X.loc[data.nutrition_grade_fr.isnull(), X.columns.str.startswith("nutrition_grade_fr_")] = np.nan
X.shape

# data.isnull().sum()
# data.describe(include='all')
# data['pnns_groups_1'].value_counts(dropna=False)
```

```python
# utiliser le minmaxscaler pour que les colonnes aient le même impact
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
X_scaled.head()
```

```python
# imputation
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=1) #k=1
X_imputed = imputer.fit_transform(X_scaled)
```

```python
# décoder
X_unscaled = pd.DataFrame(scaler.inverse_transform(X_imputed), columns=X_scaled.columns)
```

```python
# repasser les dummies en variable qualitatives initiales
group_one = X_unscaled.loc[:, 'pnns_groups_1_Beverages':'pnns_groups_1_sugary-snacks'].columns
group_two = X_unscaled.loc[:, 'pnns_groups_2_Alcoholic beverages':'pnns_groups_2_vegetables'].columns
nutritions_grades = X_unscaled.loc[:, 'nutrition_grade_fr_a':'nutrition_grade_fr_e'].columns

# créer les 3 colonnes
X_final = pd.concat([X_unscaled,pd.DataFrame(columns=['pnns_groups_1', 'pnns_groups_2', 'nutrition_grade_fr'])])

for col in group_one:
    X_final.loc[X_final[col] == 1, 'pnns_groups_1'] = col.replace('pnns_groups_1_', '')
    
for col in group_two:
    X_final.loc[X_final[col] == 1, 'pnns_groups_2'] = col.replace('pnns_groups_2_', '')
    
for col in nutritions_grades:
    X_final.loc[X_final[col] == 1, 'nutrition_grade_fr'] = col.replace('nutrition_grade_fr_', '')
    
X_final.drop(columns=X_unscaled.loc[:, 'pnns_groups_1_Beverages':].columns, inplace=True)
```

```python
# concat les deux dataframes
df = pd.concat([data_not_imputed,X_final], axis=1)
```
